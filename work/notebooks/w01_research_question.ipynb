## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Before running the code below, ensure you have a placeholder CSV file named customer_churn_starter.csv inside a folder named data, or update the path to point to your internship dataset. We will load the data framework to inspect total rows, missing baseline values, and our baseline target class distribution.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import numpy as np
import pandas as pd

# For demonstration safety, let's generate a temporary synthetic file if it doesn't exist yet
# This ensures your notebook runs flawlessly from top to bottom on the first go
import os
if not os.path.exists("data"):
    os.makedirs("data")

if not os.path.exists("data/customer_churn_starter.csv"):
    synthetic_data = pd.DataFrame({
        "account_age_months": np.random.randint(1, 48, size=500),
        "monthly_usage_hours": np.random.uniform(10.0, 150.0, size=500),
        "support_tickets_count": np.random.randint(0, 10, size=500),
        "is_churned": np.random.choice([0, 1], size=500, p=[0.8, 0.2])
    })
    synthetic_data.to_csv("data/customer_churn_starter.csv", index=False)

# --- Humanized Data Loading & Extraction Pipeline ---

target_data_path = "data/customer_churn_starter.csv"
loaded_customer_dataframe = pd.read_csv(target_data_path)

# Extract core, high-level metrics from the data frame matrix
total_recorded_observations = len(loaded_customer_dataframe)
total_missing_values_count = loaded_customer_dataframe.isnull().sum().sum()

# Determine the exact ratio of churned vs retained entries
churn_column_series = loaded_customer_dataframe["is_churned"]
total_churned_cases = int(churn_column_series.sum())
overall_churn_proportion = (total_churned_cases / total_recorded_observations) * 100

# Present our numbers with explicit, easily scannable descriptions
print("--- Exploratory Data Analysis Sanity Check ---")
print(f"Total Rows Processed:           {total_recorded_observations}")
print(f"Total Missing Data Points:      {total_missing_values_count}")
print(f"Active Historical Churn Count:  {total_churned_cases} accounts")
print(f"Baseline System Churn Rate:     {overall_churn_proportion:.2f}%")
```
